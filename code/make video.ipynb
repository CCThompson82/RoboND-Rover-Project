```python
"""Get list of images to make a video"""
imgs_path = os.path.join('data', '20171103_run1')


fnames = os.listdir(imgs_path)
img_path_list = [os.path.join(imgs_path, f) for f in fnames]
img_path_list.sort()
print(img_path_list[:3])
```

```python
"""Setup"""
# Import everything needed to edit/save/watch video clips
from moviepy.editor import VideoFileClip
from moviepy.editor import ImageSequenceClip


# Define pathname to save the output video
output = 'output/submission_run1.mp4'
clip = ImageSequenceClip(img_path_list, fps=60) # Note: output video will be sped up because 
                                          # recording rate in simulator is fps=25
    
```
